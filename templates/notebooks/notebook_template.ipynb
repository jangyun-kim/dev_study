## 3. Sample Feature Engineering
이 Notebook 템플릿은 기본 Feature Engineering 예시를 포함.

실제 프로젝트에서는 이 부분을 자유롭게 수정.
----

```python
# Feature Engineering Sample
if df is not None:
    # datetime 변환 예시
    if "event_time" in df.columns:
        df["event_time"] = pd.to_datetime(df["event_time"], errors="coerce")

    # 문자열 길이
    if "event_text" in df.columns:
        df["text_len"] = df["event_text"].astype(str).apply(len)

    # 시간 기반 파생변수
    if "event_time" in df.columns:
        df["hour"] = df["event_time"].dt.hour

    display(df.head())
else:
    print("⚠️ df is None.")
```
